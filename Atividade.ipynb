```python
from os import system


pessoas =  {"vini": "123",  "jonas": "qwe"}


usuarios = {"usuarioLogado": "", "senhaLogado": "", "dinheiro": 0, "bloqueado": False}

listaGerenciarProdutos = {}

def inicializarCaixa():


    print("------- inicializa√ß√£o de caixa -------")
    usuario = input("Digite o seu usu√°rio:")
    senha = input("Digite a sua senha:")

    if (pessoas["vini"] == senha or pessoas["jonas"] == senha):
        usuarios["usuarioLogado"] = usuario
        usuarios["senhaLogado"] = senha
        usuarios["dinheiro"] = int(input("Digite a quantidade de dinheiro disponivel:"))
        print(" -------- Caixa inicializado -------- ")
    else:
        print("‚ö† Seu usu√°rio ou senha est√£o errados")

    opcao = input("Digite (V) para voltar")
    if opcao == "v" or opcao == "V":
        limpar()

def limpar():
    system("cls")

limpar()


while True:

    if usuarios["bloqueado"] == True:
        print("### üîê Seu caixa est√° bloqueado ####")
        senha = input("digite a senha do usu√°rio para desbloquear")
        if senha == usuarios["senhaLogado"]:
            usuarios["bloqueado"] = False

    else:
        print('‚úÖ1 - Inicializar caixa')
        print("üçé 2 - Gerenciar Produtos")
        print("üí∞ 3 - Passar Compras")
        print("‚ùå 4 - Bloqueio de Caixa")
        print("üîê 5 - Fechar Caixa")
        print("üî¥ 6 - Sair do Sistema")

        opcaoSelecionada = int(input("Qual √© sua escolha:"))

        if opcaoSelecionada == 1:
           inicializarCaixa()

        elif opcaoSelecionada == 2:
            print("------- Gerenciar Produtos -------")

            while True:
                print("Digite 1 para adicionar um produto: ")
                print("Digite 2 para alterar ")
                print("Digite 3 para excluir ")
                print("Digite 4 visualizar ")
                print("Digite 5 pesquisar ")
                print("Digite 6 para voltar ")

                op = int(input("Digite uma op√ß√£o: "))


                if op == 1:
                    nomeProduto = input("Digite o produto que quer adicionar: ")
                    preco = float(input("Digite o pre√ßo do produto: "))
                    listaGerenciarProdutos[nomeProduto] = preco

                if op == 2:
                  paraAlterar = input("Digite o item contido na lista para alterar: ")
                  novoPreco = float(input("Digite o novo pre√ßo: "))
                  listaGerenciarProdutos[paraAlterar] = novoPreco
                  nomeLista = input("Digite um novo produto: ")
                  listaGerenciarProdutos[nomeLista] = listaGerenciarProdutos.pop(paraAlterar)

                  print(listaGerenciarProdutos)

                if op == 3:
                    paraExcluir = input("Digite o item contido na lista para excluir: ")
                    del listaGerenciarProdutos[paraExcluir]

                    print(listaGerenciarProdutos)

                if op == 4:
                    for produto, preco in listaGerenciarProdutos.items():
                        print(f"{produto}: R${preco}")

                if op == 5:
                    paraPesquisar = input("Digite o produto que voc√™ quer pesquisar: ")

                    if paraPesquisar in listaGerenciarProdutos.keys():
                        print(f"Produto: {paraPesquisar}, Valor: {listaGerenciarProdutos[paraPesquisar]}")
                    else:
                        print("Produto n√£o encontrado.")

                if op == 6:
                    break

        elif opcaoSelecionada == 3:

            print("------- Passar Compras -------")

            if not listaGerenciarProdutos:
                print("Aten√ß√£o: N√£o h√° produtos dispon√≠veis.")
                #return


            carrinho = {}
            while True:
                print("Produtos dispon√≠veis:")
                for produto, preco in listaGerenciarProdutos.items():
                    print(f"{produto}: R${preco}")

                produto_compra = input("Digite o nome do produto que deseja comprar (ou 'sair' para finalizar): ")

                if produto_compra.lower() == 'sair':
                    break

                if produto_compra in listaGerenciarProdutos:
                    quantidade = int(input("Digite a quantidade: "))
                    if produto_compra in carrinho:
                        carrinho[produto_compra] += quantidade
                    else:
                        carrinho[produto_compra] = quantidade
                    print(f"{quantidade} unidades de {produto_compra} adicionadas ao carrinho.")
                else:
                    print("Produto n√£o encontrado. Tente novamente.")



        elif opcaoSelecionada == 4:
                print("------- Bloqueio do caixa -------")
                resposta = input("Tem certeza que deseja bloquear o caixa? (S/N)")
                if resposta.lower() == "s":
                   limpar()



        elif opcaoSelecionada == 5:
                print("------- Fechar caixa ------")


        elif opcaoSelecionada == 6:
            print("------- Saindo do sistema de mercado ------")
            exit()
        else:
            print("Aten√ß√£o‚ùó: voc√™ digitou uma op√ß√£o inv√°lida ")


```
