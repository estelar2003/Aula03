```python
from os import system

produtos = []

produto1 = {"nome": "Refrigerante Dolly", "preco": 10.00}
produto2 = {"nome": "Chokito", "preco": 5.00}
produto3 = {"nome": "Caf√©", "preco": 7.50}
produto4 = {"nome": "Bala", "preco": 1.70}
produto5 = {"nome": "Donuts", "preco": 7.50}

produtos.append(produto1)
produtos.append(produto2)
produtos.append(produto3)
produtos.append(produto4)
produtos.append(produto5)

clientes = []

carrinho = []
registroDeVendas = []


def limpar():
    system("cls")


def fazer_pagamento():
    print("Resumo de Compra: ")
    total = 0

    for item in carrinho:
        produto = produtos[item["produto"] - 1]
        nome_produto = produto["nome"]
        preco_produto = produto["preco"]
        quantidade = item["quantidade"]
        subtotal = preco_produto * quantidade
        total += subtotal

        print(f"{nome_produto} : {preco_produto:.2f} x {quantidade} = {subtotal:.2f}")

    print(f"\nTotal: R$ {total:.2f}")

    cupom = int(input("Digite o n√∫mero do cupom: "))
    if cupom == 57:
       desconto = total * 0.10
       total_desconto = total - desconto
       print(f"\nTotal: R$ {total_desconto:.2f}")
       saldo = float(input("Digite o seu saldo: R$ "))
       if saldo < total_desconto:
           print("Saldo insuficiente‚ùå")
       else:
           print("Compra realizada com sucesso!‚úî")
    else:
        print("Cupom inv√°lido")
        saldo = float(input("Digite o seu saldo: R$ "))
        if saldo < total:
            print("Saldo insuficiente‚ùå")
        else:
            print("Compra realizada com sucesso!‚úî")


def cadastrar_cliente():
    nome = input("Digite seu nome: ")
    idade = int(input("Digite sua idade: "))
    email = input("Digite seu email: ")

    clientes.append({"nome": nome, "idade": idade, "email": email})
    print("Cadastro realizado com sucesso!‚úî")


def relatorio_clientes():
    if not clientes:
        print("N√£o h√° clientes cadastrados‚ùå")
    else:
        print("------ Relat√≥rio de Clientes ------")
        for cliente in clientes:
            print(f"Nome: {cliente['nome']}, Idade: {cliente['idade']}, Email: {cliente['email']}")
        print("-----------------------------------")

print("------ üíª Bem-vindo(a) √† M√°quina SaborExpresso! ------")
cadastro = input("Deseja fazer o seu cadastro? Digite (s) para sim e (n) para n√£o: ")
if cadastro == "s":
   cadastrar_cliente()

while True:
    print("------ üíª Bem-vindo(a) √† M√°quina SaborExpresso! ------")
    print("üßÉ 1. Refrigerante Dolly - R$ 10,00")
    print("üç´ 2. Chokito - R$ 5,00")
    print("‚òï 3. Caf√© - R$ 7,50")
    print("üç¨ 4. Bala - R$ 1,70")
    print("üç© 5. Donuts - R$ 7,50")


    opcao = int(input("Digite o produto que deseja: "))

    if opcao == 99:
        print("----------------- Menu administrativo -------------------")
        print(" 1 - Relat√≥rio de vendas")
        print(" 2 - Relat√≥rio de clientes")
        opcaoMenuAdministrativo = int(input("Escolha sua op√ß√£o: "))

        if opcaoMenuAdministrativo == 1:
            if not registroDeVendas:
                print("N√£o h√° registros")
            else:
                for item in registroDeVendas:
                    total = 0
                    produto = produtos[item["produto"] - 1]
                    nome_produto = produto["nome"]
                    preco_produto = produto["preco"]
                    quantidade = item["quantidade"]
                    subtotal = preco_produto * quantidade
                    total += subtotal

                    print(f"\n---{nome_produto} : {preco_produto:.2f} x {quantidade} = {subtotal:.2f}")
                    print(f"--Total: R$ {total:.2f}\n")



        elif opcaoMenuAdministrativo == 2:
            relatorio_clientes()

    elif 1 <= opcao <= 5:
        quantidade = int(input(f"Digite a quantidade de {produtos[opcao - 1]['nome']}: "))
        if quantidade > 10:
            print("Voc√™ s√≥ pode comprar 10 unidades de cada produto ):")
            continue

        carrinho.append({"produto": opcao, "quantidade": quantidade})
        registroDeVendas.append({"produto": opcao, "quantidade": quantidade})
        mais_produtos = input("Deseja comprar mais produtos? (s) sim (n) n√£o: ").lower()

        if mais_produtos == "n":
           fazer_pagamento()
        continue

    else:
        print("Digite uma op√ß√£o v√°lida!")

    limpar()
```
